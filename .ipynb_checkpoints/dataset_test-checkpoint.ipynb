```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("./data/dataset_trimmed.csv")
dataset = dataset.replace([0, "To demolish", "Under construction", "To restore"], np.nan)
```

```python
print(dataset.shape)
dataset.head(10)
```

```python
condition = dataset["Garden"] == False
dataset.loc[condition, "Garden Area"] = dataset.loc[condition, "Garden Area"].fillna(0)

condition = dataset["Terrace"] == False
dataset.loc[condition, "Terrace Area"] = dataset.loc[condition, "Terrace Area"].fillna(0)

condition = dataset["Type of property"] == "apartment"
sublist = ["Number of rooms"]
dataset.loc[condition, sublist] = dataset.loc[condition, sublist].fillna(1)
sublist = ["Surface of the land", "Terrace Area", "Garden Area"]
dataset.loc[condition, sublist] = dataset.loc[condition, sublist].fillna(0)
```

```python
# cleaning here
dataset = dataset.drop(["Garden", "Terrace"], axis = 1)
dataset = dataset.dropna()
dataset = dataset.rename(columns = {
    "Locality": "locality",
    "Type of property": "type",
    "Subtype of property":"subtype",
    "Price": "price",
    "Number of rooms": "rooms",
    "Living Area": "living_area",
    "Terrace Area": "terrace",
    "Garden Area": "garden",
    "Surface of the land": "land_area",
    "Number of facades": "facades",
    "State of the building": "state",
    "Furnished": "furnished",
    "Swimming pool": "pool"
})
dataset = dataset.reset_index(drop = True)
dataset = dataset.replace({
    "To renovate": 1,
    "To be renovated": 2,
    "Normal": 4,
    "Fully renovated": 6,
    "Excellent": 7,
    "New": 8
})
dataset.head(10)
```

```python
print("Data left: ", dataset.shape[0], sep = '')
```

```python
dataset = pd.get_dummies(dataset, columns=["type"])
```
